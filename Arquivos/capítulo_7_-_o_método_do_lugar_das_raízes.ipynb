# Capítulo 7 - O método do lugar das raízes

O lugar das raízes no plano pode se determinado por meio de um método gráfico. O gráfico do lugar das raízes à medida que varia um parâmetro do sistema é conhecido como o diagrama do lugar das raízes. O lugar das raízes é uma ferramente poderosa para se analsiar e projetar sistemade controle com retroação e se constitui o principal tópico deste capítulo.

## Introdução

A estabilidade relativa e o desempenho de um sistema de controle a malha feachda estão diretamente relacionados com a localização das raízes da equação característica a malha fechada no plano. Frequentemente é necessário ajustar um ou mais parâmetros a fim de se obter o posicionamente adequado das raízes. Portanto, vale a pena determinar como as raízes da equação característica de um dado sistema se deslocam pelo plano à medida que os parâmetros mudam de valor. Ou seja, é útil determinar o lugar das raízes no plano à medida que um parâmetro tenha o seu valor modificado.

O método do lugar das raízes foi introduzido por Evans em 1948 e tem sido desenvoldido e utilizado extensamente na prática da engenharia de controle. A técnica do lugar das raízes é um método gráfico de esboçar o lugar geométrico das raízes à medida que um parâmetro é variado. De fato o método do lugar das raízes fornece ao engenheiro uma medida da sensibilidade das raízes do sistema a uma variação de um parâmetro sob consideração. A técnica do lugar das raízes pode ser usada com grande vantagem em conjunto com o critério de Routh-Hurwitz.
